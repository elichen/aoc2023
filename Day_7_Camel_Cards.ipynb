```python
input = [x.rstrip() for x in open("input.txt").readlines()]
```

```python
from collections import Counter

def hand_type(hand):
    # Count the frequency of each card in the hand
    counts = Counter(hand)
    freqs = sorted(counts.values(), reverse=True)

    # Determine the type of hand based on the frequency of cards
    if freqs == [5]:
        return 7  # Five of a kind
    elif freqs == [4, 1]:
        return 6  # Four of a kind
    elif freqs == [3, 2]:
        return 5  # Full house
    elif freqs == [3, 1, 1]:
        return 4  # Three of a kind
    elif freqs == [2, 2, 1]:
        return 3  # Two pair
    elif freqs == [2, 1, 1, 1]:
        return 2  # One pair
    else:
        return 1  # High card
```

```python
def compare_hands(hand1, hand2):
    # Define the strength order of the cards
    card_strength = {'A': 13, 'K': 12, 'Q': 11, 'J': 10, 'T': 9, '9': 8,
                     '8': 7, '7': 6, '6': 5, '5': 4, '4': 3, '3': 2, '2': 1}

    # Compare the types of the hands first
    hand1_type = hand_type(hand1)
    hand2_type = hand_type(hand2)

    if hand1_type > hand2_type:
        return -1
    elif hand1_type < hand2_type:
        return 1
    else:
        # If types are the same, compare individual cards
        for card1, card2 in zip(hand1, hand2):
            if card_strength[card1] < card_strength[card2]:
                return -1
            elif card_strength[card1] > card_strength[card2]:
                return 1
        # If all cards are the same, the hands are equal
        return 0
```

```python
def card_compare(left, right):
  hand1, hand2 = left[0], right[0]
  t1 = hand_type(hand1)
  t2 = hand_type(hand2)
  if t1 == t2: return compare_hands(hand1, hand2)
  elif t1 < t2: return -1
  else: return 1
```

```python
from functools import cmp_to_key

sorted_cards = sorted([tuple(line.split()) for line in input], key=cmp_to_key(card_compare))
```

```python
total = 0
for i,t in enumerate(sorted_cards):
  total += (i+1)*int(t[1])
total
```

```python
def card_compare2(left, right):
  hand1, hand2 = left[0], right[0]
  t1 = hand_type2(hand1)
  t2 = hand_type2(hand2)
  if t1 == t2: return compare_hands2(hand1, hand2)
  elif t1 < t2: return -1
  else: return 1
```
