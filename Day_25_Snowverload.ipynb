```python
data = """
jqt: rhn xhk nvd
rsh: frs pzl lsr
xhk: hfx
cmg: qnr nvd lhk bvb
rhn: xhk bvb hfx
bvb: xhk hfx
pzl: lsr hfx nvd
qnr: nvd
ntq: jqt hfx bvb xhk
nvd: lhk
lsr: lhk
rzs: qnr cmg lsr rsh
frs: qnr lhk lsr
""".split('\n')
```

```python
data = open("input.txt").readlines()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Parse data
graph_data = {}
for line in data:
    parts = line.split(": ")
    node, edges = parts[0], parts[1].split()
    graph_data[node] = edges

# Create a graph
G = nx.Graph()
for node, edges in graph_data.items():
    for edge in edges:
        G.add_edge(node, edge)

# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color='lightblue', font_size=10, node_size=2000, edge_color='gray')
plt.title("Bidirectional Graph Visualization")
plt.show()
```

```python
edges = [("xkf", "rcn"), ("dht", "xmv"), ("thk", "cms")]

for e in edges:
  G.remove_edge(*e)

for subgraph in nx.connected_components(G):
  print(len(subgraph))
```
